- 正弦波`圖表`範例


```python
import numpy as np
import matplotlib.pyplot as plt

# 生成 x 軸數據 (0 到 2π 之間的 100 個點)
x = np.linspace(0, 2 * np.pi, 100)

# 計算正弦函數值
y = np.sin(x)

# 繪製圖形
plt.plot(x, y, label='sin(x)', color='b')

# 添加標題與標籤
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.axhline(0, color='black', linewidth=0.5)  # 添加 x 軸
plt.axvline(0, color='black', linewidth=0.5)  # 添加 y 軸
plt.grid(True, linestyle='--', alpha=0.6)  # 添加網格線
plt.legend()  # 顯示圖例

# 顯示圖形
plt.show()
```
